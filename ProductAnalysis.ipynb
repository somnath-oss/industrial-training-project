```python
#We can also see that the column memory is of float data type, but it is actually a int.
#We can convert it to a f using the following code:
dataset['memory_size'] = dataset['memory_size'].astype(int)
```

```python
# drop all duplicate values in rating category

dataset['battery_size'].dropna(inplace=True)
```

```python
#drop all duplicate values from dataset
dataset.dropna(inplace=True)
```

```python
# check for missing values

dataset.isnull().sum()
```

```python
# the distribution of best_price
plt.title("count vs best price")
sns.countplot(x='best_price', data=dataset)
```

```python
# what was the best price of brand
plt.title("best price of model")
dataset.groupby('brand_name')['best_price'].count().plot(kind='bar')
plt.ylabel("best price")
```

```python
# which brand sold their product in lowest price
plt.title("lowest price of brand")
dataset.groupby('brand_name')['lowest_price'].count().sort_values(ascending=False).head(10).plot(kind='bar')
plt.ylabel("lowest price")
```

```python
# category best price of model
plt.title("model name vs best price",color="darkblue", family="serif",size=20)
dataset.groupby('model_name')['best_price'].count().sort_values(ascending=False).head(10).plot(kind='pie', autopct='%1.1f%%')
```

```python
# category memory size for device models
plt.title("brand name vs memory",color="g", family="serif",size=20)
dataset.groupby('brand_name')['memory_size'].count().sort_values(ascending=False).head(10).plot(kind='pie', autopct='%1.1f%%')
```

** We can see that the brand name of Samsung, Apple and Xiaomi had the most sales**

```python
#configure the poopularity of os using boxplot method
sns.boxplot(x='os', y='popularity', data=dataset, palette='flare', width=0.8)
plt.xlabel('OS')
plt.ylabel('Popularity')
plt.title('Popularity of OS', color='darkblue')
plt.figure(figsize=(12,6))
plt.show()
```

```python
#check best selling brand
dataset.groupby('brand_name')['sellers_amount'].count().sort_values(ascending=False).head(20).plot(kind='bar', color='deeppink')
plt.xlabel('Brands')
plt.ylabel('Sellers Amount')
plt.title('The Best-Selling Brand')
plt.gcf().set_size_inches(8,5)
plt.show()
```

```python
# check mobile's memory
sns.countplot(x='memory_size', data=dataset, palette='viridis', order=dataset['memory_size'].value_counts().index)
plt.title('Count of Memory Size')
plt.figure(figsize=(10,6))
plt.show()
```

**we see that 128 and 64 GB memory are pick point and 4GB memory is least point**

```python
#popularity of the product of bands
dataset.iloc[10:90].groupby('release_date')['popularity'].count().plot(color='green', linewidth=3, markersize=10)
plt.xlabel('Popularity')
plt.ylabel('Release Date')
plt.title('Popularity Trend over Time')
plt.legend()
plt.show()
```

```python
#comarison between highest price and lowest price
sns.scatterplot(x='highest_price', y='lowest_price', data=dataset, hue='best_price')
plt.xlabel('Highest Price')
plt.ylabel('Lowest Price')
plt.title('Comparison between Highest Price and Lowest Price')
plt.show()
```

```python
#overall data analysis of dataset
dataset.head(25).plot.hist(color=['purple','deeppink','blue','skyblue','violet'], edgecolor='black', linewidth= 1.5)
plt.title('Histogram Analysis of Phone Sales Data')
plt.show()
```

```python
x = dataset['brand_name']
y = dataset['sellers_amount']
colors = np.random.randint(100, size=(len(dataset)))
size = .1* np.random.randint(100, size=(len(dataset)))

plt.scatter(x, y, c=colors, s=size, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

```python
#compair between best price vs battery size and  best price vs screen size

# best price vs battery size
x_battery = dataset['best_price']
y_battery = dataset['battery_size']
plt.scatter(x_battery, y_battery, color='green', label='Battery Size')

# best price vs screen size
x_screen = dataset['best_price']
y_screen = dataset['screen_size']
plt.scatter(x_screen, y_screen, color='darkblue', label='Screen Size')

plt.xlabel('Best Price')
plt.ylabel('Attribute Size')  # Adjust the ylabel based on your data
plt.title('Scatter Plots of Best Price and Attribute Size')
plt.legend()
plt.show()
```
